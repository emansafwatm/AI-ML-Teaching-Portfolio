```python

# l-bfgs-b algorithm local optimization of a convex function
from scipy.optimize import minimize
from numpy.random import rand
```

Define a simple two-dimensional objective function, a bowl function, e.g. x^2. It is simple the sum of the squared input variables with an optima at f(0, 0) = 0.0.




```python
# objective function
def objective(x):
	return x[0]**2.0 + x[1]**2.0
```

Define a function for the derivative of the function, which is [x*2, y*2].

```python
# derivative of the objective function
def derivative(x):
	return [x[0] * 2, x[1] * 2]
```

define the bounds of the function as a box with the range -5 and 5 in each dimension

```python
# define range for input
r_min, r_max = -5.0, 5.0
```

The starting point of the search will be a randomly generated position in the search domain.

```python

# define the starting point as a random sample from the domain
pt = r_min + rand(2) * (r_max - r_min)
```

Apply the BFGS algorithm to find the minima of the objective function by specifying the name of the objective function, the initial point, the method we want to use (BFGS), and the name of the derivative function.

```python

# perform the bfgs algorithm search
result = minimize(objective, pt, method='BFGS', jac=derivative)
```

review the result reporting a message as to whether the algorithm finished successfully or not and the total number of evaluations of the objective function that were performed.

```python

# summarize the result
print('Status : %s' % result['message'])
print('Total Evaluations: %d' % result['nfev'])
```

report the input variables that were found and their evaluation against the objective function.

```python
# evaluate solution
solution = result['x']
evaluation = objective(solution)
print('Solution: f(%s) = %.5f' % (solution, evaluation))
```
