# Root Finding with the Bisection Method

This notebook demonstrates the bisection method for approximating a real root of the nonlinear equation:

\[
f(x) = x^3 - 5x - 9
\]

## Method Overview

The bisection method requires two initial values, \(x_0\) and \(x_1\), such that:

\[
f(x_0)f(x_1) < 0
\]

A sign change indicates that at least one root lies between the two endpoints, assuming the function is continuous on that interval.

At each iteration, the midpoint is calculated as:

\[
x_2 = \frac{x_0 + x_1}{2}
\]

The algorithm then checks which half of the interval contains the sign change:

- if \(f(x_0)f(x_2) < 0\), the new interval is \([x_0, x_2]\);
- otherwise, the new interval is \([x_2, x_1]\).

The interval is repeatedly reduced until the function value at the midpoint satisfies the requested tolerance.

## Inputs

The program requests:

- `First Guess`: lower or first interval endpoint;
- `Second Guess`: upper or second interval endpoint;
- `Tolerable Error`: acceptable value of \(|f(x_2)|\).

For example, the interval `[2, 3]` brackets the positive root because:

\[
f(2) = -11
\]

and:

\[
f(3) = 3
\]

````python
# Defining Function
def f(x):
    return x**3-5*x-9

# Implementing Bisection Method
def bisection(x0,x1,e):
    step = 1
    print('\n\n*** BISECTION METHOD IMPLEMENTATION ***')
    condition = True
    while condition:
        x2 = (x0 + x1)/2
        print('Iteration-%d, x2 = %0.6f and f(x2) = %0.6f' % (step, x2, f(x2)))

        if f(x0) * f(x2) < 0:
            x1 = x2
        else:
            x0 = x2

        step = step + 1
        condition = abs(f(x2)) > e

    print('\nRequired Root is : %0.8f' % x2)
````

````python
# Input Section
x0 = input('First Guess: ')
x1 = input('Second Guess: ')
e = input('Tolerable Error: ')

# Converting input to float
x0 = float(x0)
x1 = float(x1)
e = float(e)

#Note: You can combine above two section like this
# x0 = float(input('First Guess: '))
# x1 = float(input('Second Guess: '))
# e = float(input('Tolerable Error: '))


# Checking Correctness of initial guess values and bisecting
if f(x0) * f(x1) > 0.0:
    print('Given guess values do not bracket the root.')
    print('Try Again with different guess values.')
else:
    bisection(x0,x1,e)
````

## Result Interpretation

For an interval such as `[2, 3]`, the algorithm repeatedly halves the interval and converges toward the positive root:

\[
x \approx 2.85519654
\]

The displayed output includes:

- the iteration number;
- the current midpoint \(x_2\);
- the value \(f(x_2)\);
- the final approximate root.

A smaller tolerable error generally requires more iterations and produces a more precise approximation.
