# Breadth-First Search on Unweighted Graphs

This notebook demonstrates Breadth-First Search (BFS) using four graph examples.

## Learning Objectives

After completing this notebook, students should be able to:

1. Represent a graph using an adjacency dictionary.
2. Explain why BFS uses a first-in, first-out queue.
3. Track visited nodes to prevent repeated processing.
4. Trace the traversal of a graph level by level.
5. Apply one reusable BFS function to multiple graphs.
6. Explain the relationship between BFS and shortest paths in unweighted graphs.

## Algorithm Overview

Breadth-First Search explores a graph one level at a time.

Starting from a selected node, the algorithm:

1. marks the start node as visited;
2. places it in a queue;
3. removes the first node from the queue;
4. visits each unvisited neighbor;
5. adds those neighbors to the end of the queue;
6. continues until the queue is empty.

A queue follows the **first-in, first-out** principle. Therefore, nodes discovered earlier are processed before nodes discovered later.

For an unweighted graph, BFS can be used to find paths containing the minimum number of edges from the start node.

```python
from collections import deque


def breadth_first_search(graph, start_node):
    """
    Traverse a graph using Breadth-First Search.

    Parameters
    ----------
    graph : dict
        Mapping from each node to an ordered collection of neighbors.
    start_node : str
        Node from which traversal begins.

    Returns
    -------
    list
        Nodes in BFS visitation order.
    """
    if start_node not in graph:
        raise KeyError(
            f"Start node {start_node!r} is not present in the graph."
        )

    visited = {start_node}
    queue = deque([start_node])
    traversal_order = []

    while queue:
        current_node = queue.popleft()
        traversal_order.append(current_node)

        for neighbor in graph.get(current_node, ()):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return traversal_order


graphs = {
    "Example 1": {
        "a": ("b", "c"),
        "b": ("a", "d"),
        "c": ("a", "d"),
        "d": ("e",),
        "e": ("a",),
    },
    "Example 2": {
        "s": ("a", "b", "c"),
        "a": ("s", "d"),
        "b": ("s", "d"),
        "c": ("s", "d"),
        "d": ("s",),
    },
    "Example 3": {
        "1": ("2", "3"),
        "2": ("1", "3", "4"),
        "3": ("1", "2", "5", "6"),
        "4": ("2", "1"),
        "5": ("3",),
        "6": ("3",),
    },
    "Example 4": {
        "1": ("2", "5", "8"),
        "2": ("1", "9"),
        "3": (),
        "4": (),
        "5": ("1",),
        "6": ("8", "7", "10"),
        "7": ("6",),
        "8": ("1", "3", "4", "6"),
        "9": ("2",),
        "10": ("6",),
    },
}

start_nodes = {
    "Example 1": "a",
    "Example 2": "s",
    "Example 3": "1",
    "Example 4": "1",
}

for example_name, graph in graphs.items():
    start_node = start_nodes[example_name]
    traversal = breadth_first_search(graph, start_node)

    print(
        f"{example_name}: "
        f"{' -> '.join(traversal)}"
    )
```

## Result Interpretation

### Example 1

Traversal begins at `a`. Nodes `b` and `c` are discovered first, followed by `d` and then `e`.

### Example 2

Traversal begins at `s`. Nodes `a`, `b`, and `c` form the first search level. Node `d` is then discovered through the neighboring nodes.

### Example 3

Traversal begins at node `1`. BFS processes nodes according to their distance, measured in graph edges, from the starting node.

### Example 4

Traversal begins at node `1` and reaches a larger graph containing multiple branches. Nodes `3` and `4` are explicitly represented as leaf nodes with no outgoing neighbors.

## Queue and Visited Set

The implementation uses:

- `deque` as the FIFO queue;
- `visited` to prevent duplicate processing;
- `traversal_order` to record the result;
- ordered tuples for neighbors so the demonstrated traversal is reproducible.

A node is marked as visited when it enters the queue rather than when it leaves the queue. This prevents the same node from being inserted multiple times.

## Complexity

For a graph with \(V\) vertices and \(E\) edges, BFS has:

\[
O(V + E)
\]

time complexity, because every vertex and edge is examined at most a constant number of times.

Its auxiliary space complexity is:

\[
O(V)
\]

for the queue, visited set, and traversal result.

## Limitations

This implementation returns the visitation order only. A shortest-path implementation would additionally store the parent of each discovered node so that a path from the start node to a selected goal could be reconstructed.
