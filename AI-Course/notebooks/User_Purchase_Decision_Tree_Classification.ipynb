```python
from pathlib import Path

import pandas as pd
from sklearn.model_selection import train_test_split

data_path = Path("../datasets/User_Data.csv")

if not data_path.exists():
    raise FileNotFoundError(
        f"Dataset not found: {data_path.resolve()}"
    )

dataset = pd.read_csv(data_path)

feature_columns = ["Age", "EstimatedSalary"]
target_column = "Purchased"

X = dataset[feature_columns]
y = dataset[target_column]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y,
)

print("Training shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(
    criterion="entropy",
    random_state=42,
)

classifier.fit(X_train, y_train)
```

```python
y_train
```

3. Predicting the test result

```python
y_pred = classifier.predict(X_test)
```

4. Test accuracy of the result (confusion matrix)

```python
#Creating the Confusion matrix
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
cm
```

5. Visualizing the training set result:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap


def plot_decision_regions(
    model,
    X_data,
    y_data,
    title,
):
    x_values = X_data.to_numpy()
    y_values = y_data.to_numpy()

    age_range = np.linspace(
        x_values[:, 0].min() - 1,
        x_values[:, 0].max() + 1,
        300,
    )

    salary_range = np.linspace(
        x_values[:, 1].min() - 5000,
        x_values[:, 1].max() + 5000,
        300,
    )

    age_grid, salary_grid = np.meshgrid(
        age_range,
        salary_range,
    )

    prediction_data = pd.DataFrame(
        {
            "Age": age_grid.ravel(),
            "EstimatedSalary": salary_grid.ravel(),
        }
    )

    predictions = model.predict(
        prediction_data
    ).reshape(age_grid.shape)

    plt.figure(figsize=(10, 7))

    plt.contourf(
        age_grid,
        salary_grid,
        predictions,
        alpha=0.35,
        cmap=ListedColormap(("purple", "green")),
    )

    colors = ["purple", "green"]

    for class_value in np.unique(y_values):
        plt.scatter(
            x_values[y_values == class_value, 0],
            x_values[y_values == class_value, 1],
            label=f"Purchased = {class_value}",
            color=colors[int(class_value)],
            edgecolor="black",
            s=45,
        )

    plt.title(title)
    plt.xlabel("Age")
    plt.ylabel("Estimated Salary")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

6. Visualizing the test set result:

```python
plot_decision_regions(
    classifier,
    X_test,
    y_test,
    "Decision Tree Classification — Test Set",
)
```
