# Loading the dataset

```python
import pandas as pd

examples = [
    {
        "text": "The Da Vinci Code is an awesome book.",
        "sentiment": 1,
    },
    {
        "text": "This book is just boring and disappointing.",
        "sentiment": 0,
    },
    {
        "text": "The story was engaging and beautifully written.",
        "sentiment": 1,
    },
    {
        "text": "The plot was confusing and the ending was weak.",
        "sentiment": 0,
    },
    {
        "text": "I enjoyed every chapter of this excellent novel.",
        "sentiment": 1,
    },
    {
        "text": "The characters were dull and unconvincing.",
        "sentiment": 0,
    },
    {
        "text": "A thoughtful and memorable reading experience.",
        "sentiment": 1,
    },
    {
        "text": "The writing felt repetitive and poorly organized.",
        "sentiment": 0,
    },
    {
        "text": "This is one of the best books I have read.",
        "sentiment": 1,
    },
    {
        "text": "I would not recommend this book.",
        "sentiment": 0,
    },
    {
        "text": "The author created an interesting and exciting story.",
        "sentiment": 1,
    },
    {
        "text": "The story was slow and difficult to finish.",
        "sentiment": 0,
    },
]

train_df = pd.DataFrame(examples)

print("Dataset shape:", train_df.shape)
train_df.head()
```

# Text preprocessing

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
feature_vector =cv.fit(train_df['text'])

features = cv.get_feature_names_out()
print(f"Total lengh of vocabulary is {len(features)}, examples are {features[:200]}")
```

```python
train_df_features =cv.transform(train_df['text'])
```

```python
train_df_features
```

```python
train_df_features.getnnz()
```

```python
train_df_features.todense()
```

```python
train_df2 = pd.DataFrame(
    train_df_features.toarray(),
    columns=features,
)

train_df2.head()
```

```python
selected_terms = [
    term
    for term in [
        "the",
        "da",
        "vinci",
        "awesome",
        "just",
        "book",
    ]
    if term in train_df2.columns
]

train_df2.loc[:1, selected_terms]
train_df2.head()
```
