## Introduction to Numpy and Image Basics

- ### Goals
 - Understand how to work with the basics of Numpy
 - Understand how to create arrays
 - Slice and index elements from arrays
 - Open and display images with Numpy

```python
# importing numpy library
import numpy as np
```

```python
# creating an array and check its type
arr = np.array([1,2,3])
print("arr: ",arr)
print("type: ",type(arr))
```

```python
# crete an array of range between given start and
# end values with a given interval
#start value - 0
#stop value - 20
#interval - 4
np.arange(0,20,4)
```

```python
# creating a multidimensional array

#create 4 x 4 matrix with all zeros
print(np.zeros(shape = (4,4)))

# create 5 x 4 matrix with all ones
print(np.ones(shape=(5,4)))
```

```python
# create 5 random integer values in the range of 0-100
a = np.random.randint(0,100,5)
a
```

```python
# finding max and min values
print(a.max())
print(a.min())
```

```python
#finding position of max and min values
print(a.argmax())
print(a.argmin())
```

```python
#reshaping an array to another shape (4 x 5 matrix)
mat = np.arange(0,20).reshape((4,5))
```

```python
mat
```

```python
# accessing 2nd row and 3rd column values
mat[1,2]
```

```python
#accessing whole second column
mat[:,1]
```
